## Configs

```python
predictions_path = configs.data_configs.save_predictions_path
```

```python
delta_back = configs.optimazation_configs.days_back_predictions
```

```python
n_days = configs.data_configs.n_days
lag_days = configs.data_configs.lags
window = configs.data_configs.window
```

```python
today = datetime.date.today()
dates_str_vector = [(today - relativedelta(days = x)).strftime("%Y-%m-%d") for x in range(delta_back)]
```

```python
filenames = next(os.walk(predictions_path), (None, None, []))[2] 
```

```python
stock_code_name = "AAPL"
```

## Transfomations

```python
dfs_ = list() 
for x in filenames:
    date_ = x[0:10]
    if date_ in dates_str_vector:
        dfs_.append(pd.read_csv(f'{predictions_path}/{x}'))
batch_predictions_csv = pd.concat(dfs_)
```

```python
stocks_codes_ = batch_predictions_csv.StockCode.unique()
```

```python
stocks_codes_
```

```python
data_history_stocks = dict()
for stock_code_name in stocks_codes_:
    data_history_stocks[stock_code_name] = data_functions.get_stock_data(stock_code = stock_code_name, n_days = n_days, window = window, lags = lag_days)
```

```python
def viz_history_predictions(data_sets, predictions, stock_code):
    
    df = data_sets[stock_code]
    stock_prediction = predictions[(predictions['Type'] == 'Forecast') & (predictions['StockCode'] == stock_code)]
    

    fig = make_subplots(rows=1, cols=1,vertical_spacing = 0.1,shared_xaxes=True,
                       subplot_titles=(
                           f'stock: {stock_code} roll window analysis: {window} days')
                       )

    fig.add_trace(go.Scatter(x=df['Date'], y=df.Close, marker_color = 'blue', name='Price'),row=1, col=1)
    fig.add_trace(go.Scatter(x=df['Date'], y=df.Close_roll_mean, marker_color = 'grey', name='roll mean' ),row=1, col=1)
    fig.add_trace(go.Scatter(x=df['Date'], y=df.lower, marker_color = 'pink', name='bound' ),row=1, col=1)
    fig.add_trace(go.Scatter(x=df['Date'], y=df.upper, marker_color = 'pink', name='bound', showlegend=False ),row=1, col=1)
    
    fig.add_trace(go.Scatter(x=stock_prediction['Date'], y=stock_prediction.stock_price, mode='markers', marker_color = 'green', name='prediction'),row=1, col=1)

    fig.update_layout(height=500, width=900, title_text=f"stock {stock_code} vizualization")
    fig.show()
```

```python
for stock_code_name in stocks_codes_:
    viz_history_predictions(data_sets = data_history_stocks, predictions = batch_predictions_csv, stock_code = stock_code_name)
```
